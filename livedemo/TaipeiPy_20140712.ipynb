```python
def get_datadate(stock_id = "2454"):
    url = "https://tw.stock.yahoo.com/d/s/major_{sid}.html".format(sid=stock_id)
    res = requests.get(url)
    tables = pd.read_html(res.text)
    datatables = [t for t in tables if t.shape == (16,8)]
    if len(datatables) > 0:
        data_date = map(int,tables[7][0][1].split(u'\u8cc7\u6599\u65e5\u671f\uff1a')[-1].split("/"))
        data_date[0] = data_date[0] + 1911
        return datetime.datetime(*data_date)
    else:
        return None
    
```
