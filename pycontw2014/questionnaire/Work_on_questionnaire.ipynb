```python
_level_mapping = {
    u'不會、沒聽過': 0,
    u'有聽過，不熟': 1,
    u'會一點': 2,
    u'熟悉、經常使用': 3
}                  
def category_to_level(cell):
    if cell is np.NaN:
        return np.NaN
    else:
        return _level_mapping[cell.decode('utf8')]
```

```python
category_to_level(df.iat[2, 19]), category_to_level(np.NaN)
```
