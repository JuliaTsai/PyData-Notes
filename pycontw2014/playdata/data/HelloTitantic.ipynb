```python
np_continous_types = [np.int64, np.int32, np.int16, np.int8, np.int0, np.int_, np.int, 
                      np.float128, np.float64, np.float32, np.float16, np.float_, np.float]
```

```python
np_discrete_types = [np.string_, np.unicode_, np.str_, np.str, np.unicode]
```

```python
def convert_to_orange_feature_type(np_data_type):
    if np_data_type in np_continous_types:
        return Orange.feature.Continuous
    elif np_data_type in np_discrete_types:
        return Orange.feature.Discrete
    else:
        raise Exception("np_data_type is not np's data type!")
        
```

```python
from time import time

import numpy as np
import pylab as pl
from matplotlib import offsetbox
from sklearn import (manifold, datasets, decomposition, ensemble, lda,
                     random_projection)
```

```python
n_samples, n_features = X.shape
n_neighbors = 30
```

```python
def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    pl.figure()
    ax = pl.subplot(111)
    for i in range(X.shape[0]):
        pl.text(X[i, 0], X[i, 1], str(digits.target[i]),
                color=pl.cm.Set1(y[i] / 10.),
                fontdict={'weight': 'bold', 'size': 9})

    if hasattr(offsetbox, 'AnnotationBbox'):
        # only print thumbnails with matplotlib > 1.0
        shown_images = np.array([[1., 1.]])  # just something big
        for i in range(digits.data.shape[0]):
            dist = np.sum((X[i] - shown_images) ** 2, 1)
            if np.min(dist) < 4e-3:
                # don't show points that are too close
                continue
            shown_images = np.r_[shown_images, [X[i]]]
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(digits.images[i], cmap=pl.cm.gray_r),
                X[i])
            ax.add_artist(imagebox)
    pl.xticks([]), pl.yticks([])
    if title is not None:
        pl.title(title)
```

```python
# Isomap projection of the digits dataset
print("Computing Isomap embedding")
t0 = time()
X_iso = manifold.Isomap(n_neighbors, n_components=5).fit_transform(X)
print("Done.")
plot_embedding(X_iso[:,:2],
               "Isomap projection of the digits (time %.2fs)" %
               (time() - t0))
```
